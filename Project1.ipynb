```
#imports all here
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```
#Begin step 1: request the url
req = requests.get('https://www.spaceweatherlive.com/en/solar-activity/top-50-solar-flares',headers = {"User-Agent":"Chrome/84.0"})
page = req.content
```

```
#extract text from page
soup = BeautifulSoup(page,'html.parser')

#use prettify
soup.prettify()

#use find to save the table as a variable
space_table = soup.find('table')
```

```
#use pandas to read the html file 
space_setup = pd.read_html(str(space_table), flavor="bs4")
space_final = pd.concat(space_setup)

#extract the data from the page
space_final.columns = ['rank', 'x_class', 'date', 'region', 'start-time', 'max-time', 'end-time', 'movie']

#print the table
space_final
```

```
#Begin step 2: tidy up by removing last column
space_final.drop(['movie'], axis = 1, inplace = True)

# Use datetime import to combine the date and each of three time columns
space_final['start'] = space_final['date'].astype(str) + ' ' + space_final['start-time'].astype(str)
space_final['max'] = space_final['date'].astype(str) + ' ' + space_final['max-time'].astype(str)
space_final['end'] = space_final['date'].astype(str) + ' ' + space_final['end-time'].astype(str)

space_final['start_datetime'] = pd.to_datetime(space_final['start'])
space_final['max_datetime'] = pd.to_datetime(space_final['max'])
space_final['end_datetime'] = pd.to_datetime(space_final['end'])

# Drop the extra not needed columns
space_final.drop(['date', 'start-time', 'max-time', 'end-time', 'start', 'max', 'end'], axis = 1, inplace = True)

# Set empty reigion as missing value of NaN *CAN BE SKIPPED BUT LEFT IN*
space_final.replace('--', 'NaN', inplace = True)

#print the table
space_final
```
